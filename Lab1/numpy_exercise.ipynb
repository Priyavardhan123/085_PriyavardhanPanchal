```python
import numpy as np
```

```python
# randomly initializing numpy arrays

narr1 = np.random.randint(1,10,(3,2))
narr2 = np.random.randint(1,10,(2,3))

print(narr1)
print(narr2)
```

```python
# matrix multiplication

mul = np.dot(narr1,narr2)
print(mul)
```

```python
# mean of first matrix

mean = np.mean(narr1)
print(mean)
```

```python
import pandas as pd

mtcars = pd.read_csv('https://gist.githubusercontent.com/ZeccaLehn/4e06d2575eb9589dbe8c365d61cb056c/raw/64f1660f38ef523b2a1a13be77b002b98665cdfe/mtcars.csv')

narr = np.array(mtcars.iloc[:,1:]) # leaving first column

# finding mean centered version by column
nparrayCentered = narr.T - np.mean(narr, axis=1) # Remove the mean for each row
nparrayCentered = nparrayCentered.T # Transpose back the result

print(nparrayCentered)
```
